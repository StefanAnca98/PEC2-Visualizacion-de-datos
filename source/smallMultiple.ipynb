```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('loan_data.csv')
# Filtrar el DataFrame
data = data[data["person_home_ownership"] != "OTHER"]
data.head(n=5)
```

```python
# Configuracin del grfico
sns.set(style="whitegrid")

# Crear el objeto FacetGrid para el "small multiple"
g = sns.FacetGrid(data, col="person_education", row="person_home_ownership", height=4, aspect=1.2)

# Mapear el scatterplot con lnea de tendencia en cada subgrfico del FacetGrid
g.map_dataframe(sns.regplot, x="loan_amnt", y="loan_int_rate", scatter_kws={'alpha':0.6}, line_kws={"color": "red"})

# Ajustar etiquetas y ttulos
g.set_axis_labels("Cuanta del Prstamo", "Tasa de Inters")
g.set_titles(col_template="{col_name}", row_template="{row_name}")
g.add_legend()

plt.savefig("Small_multiple.png", format='png', dpi=300) 

# Mostrar el grfico
plt.show()
```
